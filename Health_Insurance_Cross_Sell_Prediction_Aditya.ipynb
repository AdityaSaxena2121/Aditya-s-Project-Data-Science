## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.There is a significant difference in the average annual premium between customers who have previously been insured and those who have not.

Null Hypothesis (H0): The average annual premium is the same for customers who have previously been insured and those who have not.
Alternative Hypothesis (H1): The average annual premium is different for customers who have previously been insured and those who have not.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

import pandas as pd
from scipy.stats import ttest_ind


# Groups based on Previously_Insured
insured = df[df['Previously_Insured'] == 1]['Annual_Premium']
not_insured = df[df['Previously_Insured'] == 0]['Annual_Premium']

# Perform t-test
t_stat, p_value = ttest_ind(insured, not_insured)

print(f"T-Statistic: {t_stat}, P-Value: {p_value}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in the average annual premium.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the average annual premium.")
````

##### Which statistical test have you done to obtain P-Value?

Answer Here. To obtain the p-value for comparing the average annual premium between customers who have previously been insured and those who have not, I performed an independent samples t-test. This statistical test is appropriate when you want to compare the means of two independent groups to determine if there is a significant difference between them.

##### Why did you choose the specific statistical test?

Answer Here.The independent samples t-test was chosen for this hypothesis test because it is designed to compare the means of two independent groups.
The hypothesis is about comparing the average annual premium between two groups: customers who have previously been insured and those who have not. The t-test specifically tests whether the means of two independent groups are statistically different from each other.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here. The proportion of customers interested in the insurance offer (Response = 1) is different between males and females.
Null Hypothesis (H0): The proportion of interest in the insurance offer is the same for males and females.
Alternative Hypothesis (H1): The proportion of interest in the insurance offer is different for males and females.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

import pandas as pd
from scipy.stats import chi2_contingency


# Contingency table
contingency_table = pd.crosstab(df['Gender'], df['Response'])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2}, P-Value: {p}")

# Conclusion
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis. There is a significant difference in the proportion of interest in the insurance offer between genders.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the proportion of interest in the insurance offer between genders.")
````

##### Which statistical test have you done to obtain P-Value?

Answer Here. To test the hypothesis that the proportion of customers interested in the insurance offer (Response = 1) is different between males and females, I used the Chi-Square test for independence. This statistical test is appropriate for determining if there is a significant association between two categorical variables.

##### Why did you choose the specific statistical test?

Answer Here. The Chi-Square test for independence was chosen for this hypothesis test because it is specifically designed to determine whether there is a significant association between two categorical variables. The hypothesis involves comparing the proportions of customers interested in the insurance offer (Response = 1) between two groups based on gender (Male and Female). Both of these variables are categorical.

Alternative Hypothesis (H1): The average age of customers with a driving license is different from those without.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

import pandas as pd
from scipy.stats import ttest_ind

# Groups based on Driving_License
with_license = df[df['Driving_License'] == 1]['Age']
without_license = df[df['Driving_License'] == 0]['Age']

# Perform t-test
t_stat, p_value = ttest_ind(with_license, without_license)

print(f"T-Statistic: {t_stat}, P-Value: {p_value}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in the average age of customers with and without a driving license.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the average age of customers with and without a driving license.")
````

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# Fit the Algorithm

# Predict on the model

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define the parameter grid
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Regularization parameter
    'penalty': ['l1', 'l2']  # Regularization penalty
}

# Create a logistic regression model
model = LogisticRegression(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')

# Fit the model with GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Predict on the model with best parameters
y_pred = grid_search.predict(X_test)

# Assess Model Performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
````
